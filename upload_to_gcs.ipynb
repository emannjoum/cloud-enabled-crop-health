```python
from google.cloud import storage
import os

client = storage.Client(project="ee-emannjoum")
bucket_name = 'crop-images-bucket'
bucket = client.bucket(bucket_name)

def upload_to_gcs(local_file_path, gcs_file_path):
    blob = bucket.blob(gcs_file_path)
    blob.upload_from_filename(local_file_path)
    print(f"File {local_file_path} uploaded to {gcs_file_path}.")
```

```python
# os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/kaggle/input/service-account-key.json'

drive_folder_path = '/content/drive/MyDrive/Telangana Crop Health/Telangana Images'

for file_name in os.listdir(drive_folder_path):
    if file_name.endswith('.tif'):
        local_file_path = os.path.join(drive_folder_path, file_name)
        gcs_file_path = f'telangana/{file_name}'  # GCS path (will create "images/telangana/" folder)
        upload_to_gcs(local_file_path, gcs_file_path)
```

```python
blobs = bucket.list_blobs(prefix="telangana/")
num_files = 0
for blob in blobs:
    num_files += 1

print(f"Number of files in telangana folder: {num_files}")
```
